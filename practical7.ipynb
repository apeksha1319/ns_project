```python
from nltk import sent_tokenize
from nltk import word_tokenize
text='Rohit Sharma is the best cricket player'
# sentence tokenization
sent_tokanization= nltk.sent_tokenize(text)
print(sent_tokanization)
# word tokenization
word_tokanization= nltk.word_tokenize(text)
print(word_tokanization)
```

```python
from nltk.corpus import stopwords
stopwords=set(stopwords.words('english'))
print(stopwords)
```

```python
from nltk.stem import PorterStemmer
stem=[]
for i in word_tokanization:
    ps=PorterStemmer()
    stem_word=ps.stem(i)
    stem.append(stem_word)
print(stem)
```
